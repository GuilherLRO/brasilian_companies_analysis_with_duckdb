```python
import duckdb
import os
```

```python
conn = duckdb.connect('cnpj.db')
```

```python
tables = conn.execute("SHOW TABLES").fetchall()
tables = [t[0] for t in tables]
print(tables)
```

```python
for table in tables:
    count = conn.execute(f"SELECT COUNT(*) FROM {table}").df().iloc[0,0]
    print(f"{table}: {count}")
```
